```python
centro = geocoder.komoot('araucania').latlng
mymap = folium.Map(location=centro, zoom_start=8)
for x, row in ar.iterrows():
    p1 = geocoder.komoot(row.region).latlng
    print (p1)
    size1 = row.forest * 100
    size2 = row.reforest * 10
    marker1 = row.region + ', Reforestacion: ' + str(row.reforest)
    marker2 = row.region + ', Forestacion: ' + str(row.forest)
    
    if (row.reforest < 600):
        color = '#aa2929'
    else:
        color = '#3186cc'
    
    folium.Circle(p1, radius = size2, fill=True ,popup=marker1,color= color,fill_color=color, fill_opacity=0.3).add_to(mymap)
    #azul, reforestacion
    folium.Circle(p1, radius = size1, fill=True ,popup=marker2,color='#176610',fill_color='#176610', fill_opacity=0.3).add_to(mymap)
    #verde, forestacion
```
